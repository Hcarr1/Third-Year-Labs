```python
from torchvision import models
from torch import nn
import torch
import torchvision
```

```python
transform = torchvision.transforms.Compose([torchvision.transforms.ToTensor(), torchvision.transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225)), torchvision.transforms.Resize(224)])
trainset = torchvision.datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
alexnet_pretrained = models.alexnet(pretrained=True)

for param in alexnet_pretrained.parameters():
    param.requires_grad = False
```

```python
alexnet_pretrained.classifier[6] = nn.Linear(in_features=4096, out_features=10)
```

```python
optimizer = torch.optim.Adam(alexnet_pretrained.classifier[6].parameters(), lr=0.001)
```

```python
criterion = nn.CrossEntropyLoss()
```

```python
for i in range(10):
    for Xb, Yb in iter(trainloader):
        Xb, Xy = Xb.to(device), Yb.to(device)

        optimizer.zero_grad()
        outputs = alexnet_pretrained(Xb)
        loss = criterion(outputs, Yb)
        loss.backward()
        optimizer.step()

    print(f"epoch: {i}, loss = {loss}")
```
